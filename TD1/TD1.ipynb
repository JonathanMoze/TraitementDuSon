### Synthèse d'une sinusoïde

```python
def synthese_sin(duration, amplitude, freq, samplerate, phi0):
    n = np.arange(0,duration*samplerate) # sample
    return amplitude * np.sin(2*np.pi*freq * n/samplerate + phi0)
```

```python
# Sound synthesis
amplitude = 1
samplerate = 44100 # Hz
duration = 1 # second
freq = 440.0 # Hz
phi0 = 0

n = np.arange(0,duration*samplerate) # sample

sin = synthese_sin(duration, amplitude, freq, samplerate, phi0)
```

```python
# Indicate the calc module if we are inside colaboratory
calc.setColaboratory(True)

# plot waveform
calc.plot11(n,sin,"Sound Waveform","Time (s)","Amplitude")

# Zoom
calc.plot11(n[0:200],sin[0:200],"Sound Waveform (Detail)","Time (s)","Amplitude")
```

```python
# Generate a player
Audio(sin,rate=samplerate)
```
